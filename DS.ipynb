### 🔍 Search Operation in Arrays

One of the main advantages of arrays is that you can access any element instantly if you know its index — this takes **constant time** `O(1)`.

However, if you don’t know the index and want to find an element by its value, you need to check each element one by one until you find it.  
This process is called **Linear Search**, and it takes **`O(n)` time** in the worst case.


# If you know the index (e.g., index 2):
- This is fast and takes constant time → O(1).


````python
arr = [10, 20, 30, 40, 50]
print(arr[2])
````

# In the worst case, the value might be at the end or not present at all, so the time complexity is O(n).

````python
for i in range(len(arr)):
    if arr[i] == 40:
        print(f"Found at index {i}")

````

### 🗑️ Delete Operation in Arrays

If we want to delete an item by **index**, we can access it in **constant time** `O(1)`.

However, since arrays are stored in **continuous memory**, deleting an element from **any position except the last one** creates a "gap" (or hole).  
To fix this, all the elements after the deleted item must be **shifted one step to the left**, which takes **`O(n)` time** in the worst case.

> For example, deleting the 2nd element in a 5-item array means the last 3 elements need to move to fill the gap.


delete_list.png

````python
import numpy as np
````

````python
arr=[3,21,8,4,-9,5]
````

````python
def delete_element(arr, index):

    if index < 0 or index >= len(arr):
        print("Error: Index out of range.")
        return arr
    return np.delete(arr, index)
````

````python
delete_element(arr,2)
````

### ➕ Insert Operation in Arrays

Like deleting elements, when inserting an item into an array, we may need to **shift existing elements** to make space for the new item.

- In the **worst-case scenario**, inserting at the **beginning** of the array means shifting all the elements to the right.  
  This results in a time complexity of **`O(n)`**.

- However, if we insert at the **end** of the array (and there's space), we can simply add the new item without shifting.  
  This operation takes **`O(1)`** time.


insert_list.png

````python
def insert_element(array, index, value):

    # Check if the index is valid
    if index < 0 or index > len(array):
        raise IndexError("Index is out of bounds.")

    # Insert the value at the specified index using numpy.insert
    new_array = np.insert(array, index, value)

    return new_array
````

````python
insert_element(arr,0,1)
````

### How `set()` Works Internally in Python

The `set()` data structure in Python is implemented using a **hash table**.

#### 🔧 How it works:
- A **hash table** is a data structure that uses a **hash function** to map elements to unique positions called **buckets**.
- When you add an element to a set, Python calculates the **hash value** of that element using the built-in `hash()` function.
- This hash value determines where to store the element in memory.
- Because of this, checking whether an element exists in a set (i.e., `x in s`) is **very fast** — on average **O(1)** time.

#### 🔍 Uniqueness in Sets:
- When you call `set()` on a list (or any iterable), Python:
  - Iterates through each element.
  - Uses the hash function to check if that element is already present.
  - Only keeps **unique** elements (i.e., elements with different hash values or values that compare unequal with `==`).
  
#### Example:
```python
my_list = [1, 2, 2, 3, 1]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3}

## Singly Linked List

A singly linked list is a fundamental data structure, it consists of nodes where each node contains a data field and a reference to the next node in the linked list. The next of the last node is null, indicating the end of the list. Linked Lists support efficient insertion and deletion operations.



- In a singly linked list, each node consists of two parts: data and a pointer to the next node. This structure allows nodes to be dynamically linked together, forming a chain-like sequence.

## Doubly Linked List

- A doubly linked list is a more complex data structure than a singly linked list, but it offers several advantages. The main advantage of a doubly linked list is that it allows for efficient traversal of the list in both directions. This is because each node in the list contains a pointer to the previous node and a pointer to the next node. This allows for quick and easy insertion and deletion of nodes from the list, as well as efficient traversal of the list in both directions.

## Circular Linked List

1. Circular Singly Linked List

In Circular Singly Linked List, each node has just one pointer called the "next" pointer. The next pointer of the last node points back to the first node and this results in forming a circle. In this type of Linked list, we can only move through the list in one direction.

Representation-of-circular-linked-list.webp

2. Circular Doubly Linked List:
In circular doubly linked list, each node has two pointers prev and next, similar to doubly linked list. The prev pointer points to the previous node and the next points to the next node. Here, in addition to the last node storing the address of the first node, the first node will also store the address of the last node.

- Operations like `peek` and `isEmpty` are constant time since they only involve checking values without traversal.


### 🚀 Queue – Time Complexity Comparison

| Operation            | Array (List) | Linked List | Deque (`collections.deque`) |
|----------------------|--------------|--------------|------------------------------|
| Enqueue (add)        | O(1)*        | O(1)         | O(1)                         |
| Dequeue (remove)     | O(n)         | O(1)         | O(1)                         |
| Peek (front)         | O(1)         | O(1)         | O(1)                         |
| isEmpty              | O(1)         | O(1)         | O(1)                         |

> 📝 *Note: Array/List enqueue is amortized O(1), but can be O(n) if resizing or shifting is needed during dequeue.


### 🔄 Queue: Circular Array vs Sample Array (Python List)

| Operation               | Sample Array (List) | Circular Array |
|------------------------|---------------------|----------------|
| Enqueue (insert)       | O(1)\*              | O(1)           |
| Dequeue (remove)       | O(n)                | O(1)           |
| Peek Front             | O(1)                | O(1)           |
| Peek Rear              | O(1)                | O(1)           |
| IsFull (check full)    | Not applicable      | O(1)           |
| IsEmpty (check empty)  | O(1)                | O(1)           |

> 📝 \*Enqueue in Python list is **amortized O(1)** when using `append()`,  
> but Dequeue using `pop(0)` is **O(n)** due to shifting elements.

# Search Algorithms Comparison

This table summarizes different types of search algorithms, their time and space complexities, and use cases.

| Search Type         | Time Complexity (Best) | Time Complexity (Average) | Time Complexity (Worst) | Space Complexity | Description & Use Cases |
|---------------------|------------------------|----------------------------|--------------------------|------------------|--------------------------|
| **Linear Search**   | O(1)                   | O(n)                       | O(n)                     | O(1)             | - Use when the list is unsorted.<br>- Simple to implement. |
| **Binary Search**   | O(1)                   | O(log n)                   | O(log n)                 | O(1)             | - Use on sorted arrays.<br>- Efficient for large datasets. |
| **Jump Search**     | O(1)                   | O(√n)                      | O(√n)                    | O(1)             | - Suitable for sorted arrays.<br>- Faster than linear, less than binary. |
| **Exponential Search** | O(1)               | O(log i)                   | O(log i)                 | O(1)             | - Efficient for unbounded or infinite size arrays. |
| **Interpolation Search** | O(1)             | O(log log n)               | O(n)                     | O(1)             | - Works best with uniformly distributed data. |
| **Ternary Search**  | O(1)                   | O(log₃ n)                  | O(log₃ n)                | O(1)             | - Similar to binary but divides into 3 parts.<br>- Used in unimodal functions. |
| **Fibonacci Search**| O(log n)               | O(log n)                   | O(log n)                 | O(1)             | - Uses Fibonacci numbers to split the search range.<br>- For sorted arrays. |
| **Tree Search**     | O(log n)               | O(log n)                   | O(n)                     | O(n)             | - Use in BST or balanced trees (e.g. AVL, Red-Black Tree). |
| **Hashing (Search)**| O(1)                   | O(1)                       | O(n) (in case of collisions) | O(n)         | - Used in hash tables for constant-time lookup. |
| **Depth-First Search (DFS)** | -          | -                          | -                        | O(h) or O(n)     | - Graph/tree traversal.<br>- Use when deep paths are prioritized. |
| **Breadth-First Search (BFS)** | -        | -                          | -                        | O(w) or O(n)     | - Graph/tree traversal.<br>- Finds the shortest path in unweighted graphs. |

> **Note:**
> - `n` = number of elements
> - `h` = height of the tree
> - `w` = width of the graph/tree
> - DFS and BFS are traversal/search techniques in graphs/trees, not used for arrays.

# When to Use Which Search
- **Unsorted list?** → Linear Search
- **Sorted array?** → Binary, Jump, Interpolation
- **Searching in a tree?** → Tree Search, DFS, BFS
- **Hash-based data?** → Hashing
- **Function optimization?** → Ternary Search
